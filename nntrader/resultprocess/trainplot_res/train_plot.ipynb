```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
```

```python
logmean76 = pd.read_csv("76_logmean.csv")
logmean81 = pd.read_csv("81_logmean.csv")
logmeanfree76 = pd.read_csv("76_logmeanfree.csv")
logmeanfree81 = pd.read_csv("81_logmeanfree.csv")
```

```python
index = logmean76["Step"].as_matrix()
logmean = np.stack([logmean76["Value"].as_matrix(),logmean81["Value"].as_matrix()])
logmeanfree = np.stack([logmeanfree76["Value"].as_matrix(),logmeanfree81["Value"].as_matrix()])
logmean
```

```python
fig1, ax1 = plt.subplots()
ax1.plot(index, logmean[0], label="reward_without_commission")
ax1.plot(index, logmean[1], label="reward_with_commission")
ax1.legend(loc="lower right")
plt.title("log mean")
ticks_x = ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x/1000)+"k")
ax1.xaxis.set_major_formatter(ticks_x)
ax1.set_xlabel("steps")
ax1.set_ylabel("log mean")

fig2, ax2 = plt.subplots()
ax2.plot(index, logmeanfree[0], label="reward_without_commission")
ax2.plot(index, logmeanfree[1], label="reward_with_commission")
ax2.legend(loc="lower right")
plt.title("log mean without commission fee")
ticks_x = ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x/1000)+"k")
ax2.xaxis.set_major_formatter(ticks_x)
ax2.set_xlabel("steps")
ax2.set_ylabel("log mean without commission")

plt.show()
```
